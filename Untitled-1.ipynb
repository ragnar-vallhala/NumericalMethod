```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-100,100,200*10)
```

```python
f = x*x-5*x-3000
```

```python
def fx(x):
    return x*x-5*x-3000
```

```python
plt.plot(x,f)
```

```python
x1,x2 = -50,20
i=0
while i<100:
    i+=1
    x = (x1+x2)/2
    f3 = fx(x)
    f1 = fx(x1)
    f2 = fx(x2)

    if abs(f3)<0.01 or abs(x1-x2)<0.0001:
        print("Root is ",x," ",f3)
        break
   

    if f1*f3<0:

        x2=x
    else:
        x1=x
    

    print(f3," ",x," iterations ",i)
        
```
