```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.sin(x)
```

```python
def iter(f,guess,omega=1,tol=0.01,all=[],iteration=0,limit=0):
    
    gnew=0

    if iteration==0:
        all=[]

    if limit!=0 and limit<iteration:
        print("Iteration reached")
        print("Root:",guess)
        return all,iteration


    all.append(guess)
    iteration+=1


    
    gnew = guess + omega*f(guess)
    
    if (limit==0 or tol!=0.01) and abs(gnew-guess)<tol:
        print("Tol reached")
        print("Root:",gnew)
        return all,iteration
    else:
        return iter(f,gnew,omega=omega,all=all,tol=tol,iteration=iteration,limit=limit)
```

```python
all = iter(f,0.9,limit=9)
```

```python
plt.scatter([i for i in range(len(all[0]))],all[0])
```
