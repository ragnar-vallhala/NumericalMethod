```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(-1,1,0.001) 
```

```python
b = a**2-5.6*a+0.5
```

```python
def f(x):
    return x/np.sin(x)-2*np.square(np.cos(x))
```

```python
plt.plot(a,b)
```

```python
x1= 0.5
x2 = 1
```

```python
x = x1-(f(x1)*(x1-x2))/(f(x1)-f(x2))
```

```python
plt.plot(a,b)
plt.plot((x1,x2),(f(x1),f(x2)))

plt.plot((-1,1),(0,0),color="#000")
plt.plot(x,0,"*",color="red")
```

```python
def falsePosition(f,x1,x2,tolerance,points=[x1]):
    x = x1-(f(x1)*(x1-x2))/(f(x1)-f(x2))
    print(points)
    if abs(f(x))<tolerance:
        print("root:",x)
        return points
    else:
        x2=x
        points.append(x)
        return falsePosition(f,x1,x2,tolerance,points)
```

```python

pp = falsePosition(f,-1,0.5,0.001)
```

```python
pp
```

```python
plt.plot(a,b)
plt.scatter(pp,[0 for i in range(len(pp))], color = "#ff0000")
for i in range(len(pp)-1):
    plt.plot((pp[i],pp[i+1]),(f(pp[i]),f(pp[i+1])))
plt.plot((-1,1),(0,0),color="#000")
plt.axvline(pp[-1])
```
